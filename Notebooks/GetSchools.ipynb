```python
url = f"{baseWondeUrl}/schools"
df = getWondeAPI(url, wondeHeaders, school_schema)
```

```python
dfFinal = (df.select(
            col("id").alias("school_id"),
            col("name"),
            col("establishment_number"),
            col("urn"),
            col("phase_of_education"),
            col("la_code"),
            col("mis"),
            col("address.address_postcode"))
        )
```

```python
(dfFinal.write
    .mode("overwrite")
    .option("mergeSchema", "true")
    .format("delta")
    .save(BronzeBasePath + "/Schools")
    )
```
