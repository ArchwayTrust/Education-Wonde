# Main function for calling Wonde API

```python
def getWondeAPI (apiUrl, requestHeaders, wondeSchema):
    session = requests.Session()

    #get_contents returns an iterable set of dataframes. 
    def get_contents():
        url = apiUrl
        morePages = True

        while ( morePages ):
            try:
                page = session.get(url, headers=requestHeaders).json()
            except:
                print("Something went wrong... will implement some error handling here")
            responseContents = page["data"]
            url = page["meta"]["pagination"]["next"]
            morePages = page["meta"]["pagination"]["more"]
            yield spark.createDataFrame( responseContents, schema = wondeSchema)

    # Create empty Spark dataframe
    emp_RDD = spark.sparkContext.emptyRDD()
    df = spark.createDataFrame(emp_RDD, schema = wondeSchema)

    # Iterate through pages from get_contents
    for page in get_contents():
        df = df.union(page)

    return df

 
```
