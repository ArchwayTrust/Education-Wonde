```python
import logging

logging.basicConfig(level=logging.INFO, 
                    format='%(asctime)s - %(levelname)s - %(message)s', 
                    datefmt='%d-%b-%y %H:%M:%S')
```

```python
def getWondeAPI(apiUrl, requestHeaders, wondeSchema):
    session = requests.Session()

    def get_contents():
        url = apiUrl
        more_pages = True

        while more_pages:
            try:
                page = session.get(url, headers=requestHeaders).json()
                yield spark.createDataFrame(page["data"], schema=wondeSchema)

                # Pagination management
                pagination_data = page.get("meta", {}).get("pagination", {})
                url = pagination_data.get("next")
                more_pages = pagination_data.get("more", False)

            except Exception as e:
                logging.error("Unable to retrieve response: %s", e)
                more_pages = False

    # Create an empty DataFrame
    df = spark.createDataFrame(spark.sparkContext.emptyRDD(), schema=wondeSchema)

    # Iterate through pages from get_contents
    for page in get_contents():
        df = df.union(page)

    return df
```
